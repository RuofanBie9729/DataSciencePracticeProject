##### Historicl Growth of Electric Vehicle
Overall, we are interested in the gross growth of electric vehicle in each year. The we will check whethere there is any trend in different electric vehicle type, CAFV eligibiility, make and model. After analyzing the gross market, we will study the segment market in states.

```python
##### Gross growth of registered electric vehicles
EV_data = EV[["VIN", "County", "City", "State", "Model Year", "Make", "Model", "EV Type", "CAFV", "Electric Range"]].dropna()
EV_gross_growth = EV_data.groupby("Model Year").size().to_frame("number of EV")
import matplotlib.pyplot as plt
EV_gross_growth.pivot_table(index = "Model Year", values = "number of EV").plot()
for k, v in EV_gross_growth.loc[[2010, 2013, 2018, 2023]].iterrows():
    plt.text(k, v["number of EV"], (k, v["number of EV"]))
```

We can see that the number of national registered electric vehicle has a rise in 2010 and the increase rate is also increasing over the years. During this increase, we see three drops in 2013, 2018, 2023. The most recent drop might due to the incomplete data collection in year 2024. The other two drops in 2013 and 2018 worths an investigation. In following analysis on segment market, we can pay more attentions to these two years and check whether there is such a drop in all segment markets or just in part of the segment markets.

```python
##### Gross growth of different electric vehicle types, CAFV, make and model.
### Due to the extreme low number of registered EV before year 2011, we will truncate the plot before year 2011.
EV_type_growth = EV.loc[EV["Model Year"]>2010,].groupby(["Model Year", "EV Type"]).size().to_frame("number of EV")
EV_type_growth.pivot_table(index = "Model Year", columns = "EV Type", values = "number of EV").plot.bar(stacked = True)
```

```python
EV_type_growth.pivot_table(index = "Model Year", columns = "EV Type", values = "number of EV").plot()
```

From the stacked bar plot and the line chart, we can see that over all the registered Battery EV (BEV) is more than registered Plug-in Hybrid EV (PHEV) in every year. The trend of the two types of EV are the same across the 12 years.

```python
EV_CAFV_growth = EV.loc[EV["Model Year"] > 2010,].groupby(["Model Year", "CAFV"]).size().to_frame("number of EV")
EV_CAFV_growth.pivot_table(index = "Model Year", columns = "CAFV", values = "number of EV").plot.bar(stacked = True)
```

```python
EV_CAFV_growth.pivot_table(index = "Model Year", columns = "CAFV", values = "number of EV").plot()
```

We can see that after year 2010, there are more clean alternative fuel vehicle than non-eligible EVs. The number of clean alternative fuel vehicle inceases from year 2011 to year 2018 ang then decreases to a lower level from year 2018 to 2022. We also notice that from year 2019, there exists unknow eligibility EVs due to the unresearched battery range and the number of this types of EVs are increasing quickly from year 2020 to 2023. This might due to the presence of new types of EVs. We can check this assumption in following figures analyzing the trend of EV Makes and Models over years.

```python
EV_make_growth = EV.loc[EV["Model Year"]>2010].groupby(["Model Year", "Make"]).size().to_frame("number of EV").sort_values(["Model Year", "number of EV"], ascending = False).reset_index()
EV_top3make_growth = EV_make_growth.groupby("Model Year").head(3)
EV_top3make_growth.pivot_table(index = "Model Year", columns = "Make", values = "number of EV").plot.bar(stacked = True).legend(bbox_to_anchor = (1.0, 1.0))
```

From this bar plot, we can see that from year 2018 to year 2023, Tesla dominates the market and the demond of Tesla is increasing over the years. Other than Tesla, Chevrolet, Nissan and Ford are popular EV makers. In year 2024, BMW starts to become popular choice in EV purchase. 

```python
EV_model_growth = EV.loc[EV["Model Year"] > 2010,].groupby(["Model Year", "Model"]).size().to_frame("number of EVs").sort_values(["Model Year", "number of EVs"], ascending=False).reset_index()
EV_top3model_growth = EV_model_growth.groupby("Model Year").head(3).reset_index()
EV_top3model_growth.pivot_table(index = "Model Year", columns = "Model", values = "number of EVs").plot(kind = "bar", stacked=True).legend(bbox_to_anchor = (1.0, 1.0))
```

This bar plot provides more details to previous make bar plot. We can notice that Model 3 are very popular EV since 2018. However, model Y becomes more popular than model 3 since 2021 and the popularity of model Y increases quickly from 2021 to 2023. This is probably due to the $7500 tax credit policy on electric vehicles.

```python
##### Analysis on segment markets 
### define regions
NorthEast = ["ME", "MA", "RI", "CT", "NJ", "PA", "NY", "VT", "NH"]
MidWest = ["ND", "MN", "WI", "MI", "OH", "IN", "IL", "IA", "MO", "SD", "NE", "KS"]
South = ["DE", "MD", "DC", "WV", "VA", "NC", "SC", "GA", "FL", "KY", "TN", "AL", "MS", "LA", "AR", "OK", "TX"]
West = ["WA", "OR", "CA", "AK", "MT", "ID", "WY", "NV", "UT", "CO", "AZ", "NM", "HI"]
EV_NE = EV.loc[EV["State"].isin(NorthEast)]
EV_MW = EV.loc[EV["State"].isin(MidWest)]
EV_S = EV.loc[EV["State"].isin(South)]
EV_W = EV.loc[EV["State"].isin(West)]
```

```python
EV_NE_gross_growth = EV_NE.loc[EV_NE["Model Year"] > 2010].groupby("Model Year").size().to_frame("number of EVs").reset_index()
EV_MW_gross_growth = EV_MW.loc[EV_MW["Model Year"] > 2010].groupby("Model Year").size().to_frame("number of EVs").reset_index()
EV_S_gross_growth = EV_S.loc[EV_S["Model Year"] > 2010].groupby("Model Year").size().to_frame("number of EVs").reset_index()
EV_W_gross_growth = EV_W.loc[EV_W["Model Year"] > 2010].groupby("Model Year").size().to_frame("number of EVs").reset_index()
fig = plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.xlim(2010, 2025)
plt.plot(EV_NE_gross_growth["Model Year"], EV_NE_gross_growth["number of EVs"])
plt.title("EV growth in NorthEast Market")
plt.subplot(2,2,2)
plt.xlim(2010, 2025)
plt.plot(EV_MW_gross_growth["Model Year"], EV_MW_gross_growth["number of EVs"])
plt.title("EV growth in MidWest Market")
plt.subplot(2,2,3)
plt.xlim(2010, 2025)
plt.plot(EV_S_gross_growth["Model Year"], EV_S_gross_growth["number of EVs"])
plt.title("EV growth in South Market")
plt.subplot(2,2,4)
plt.xlim(2010, 2025)
plt.plot(EV_W_gross_growth["Model Year"], EV_W_gross_growth["number of EVs"])
plt.title("EV growth in West Market")
plt.show()
```

By ploting thr growth of EVs in different segment market, we can see that over the years, most registered EVs are in the West market. Although, the number of registered EVs in other segment market are much smaller than the number of registered EVs in the West market, the number of EVs are increasing in all segment markets.

```python
##### Make prediction
import seaborn as sns
sns.set()
plt.ylabel('number of EVs')
plt.xlabel('Model Year')
plt.xticks(rotation=45)
plt.plot(EV_gross_growth.index, EV_gross_growth['number of EV'], )
```

```python
train = EV_gross_growth[EV_gross_growth.index < 2018]
test = EV_gross_growth[(EV_gross_growth.index >=2018) & (EV_gross_growth.index < 2024)]
from statsmodels.tsa.arima.model import ARIMA
y = train["number of EV"]
ARIMAmodel = ARIMA(y, order = (1,4,1))
ARIMAmodel = ARIMAmodel.fit()
y_pred = ARIMAmodel.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = ARIMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
```

```python
plt.plot(train, color = "black")
plt.plot(test, color = "red")
plt.ylabel('number of EV')
plt.xlabel('Model Year')
plt.xticks(rotation=45)
plt.title("Train/Test split for EV")
plt.plot(y_pred_out, color='green', label = 'Predictions')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

arma_rmse = np.sqrt(mean_squared_error(test["number of EV"].values, y_pred_df["Predictions"]))
print("RMSE: ",arma_rmse)
```

```python
#### Use curve_fit for prediction
from scipy.optimize import curve_fit
import numpy as np
def expgrowth(x, a, b):
    return a * np.exp(b * x)

x_data = EV_gross_growth[EV_gross_growth.index < 2024].index - min(EV_gross_growth[EV_gross_growth.index < 2024].index)
y_data = EV_gross_growth.loc[EV_gross_growth.index < 2024].values.T[0]
params, covariance = curve_fit(expgrowth, x_data, y_data)
```

```python
forecast_years = np.arange(2024, 2030) - min(EV_gross_growth[EV_gross_growth.index < 2024].index)
forecast_values = expgrowth(forecast_years, *params)
print(forecast_values)
```

The prediction results from time series model and the exponential curve are at similar level. However, the exponential curve prediction has a larger incresing rate than the time series results.
